```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv("Customers.csv")
```

```python
customers
```

```python
customers.drop(['CustomerID'], axis=1, inplace=True)
```

```python
customers.describe()
```

```python
#各年齡層分布的人數
plt.hist(x = customers['Age']);
```

```python
#Number of MALE and FEMALE Workers
sns.countplot(x = customers['Gender']);
```

```python
#年收入的分布
plt.hist(x = customers['Annual Income ($)']);
```

```python
#年收入為 0
customers[customers['Annual Income ($)'] == 0]
```

```python
#平均年收入
customers['Annual Income ($)'].mean()
```

```python
customers.loc[customers['Annual Income ($)'] == 0, 'Annual Income ($)'] = 110731.8215
```

```python
#年收入為平均年收入
customers.loc[customers['Annual Income ($)'] == 110731.8215]
```

```python
#職業為Artist的人
customers.loc[customers['Profession'].isin(['Artist'])] 
```

```python
#職業為Healthcare的人
customers.loc[customers['Profession'].isin(['Healthcare'])] 
```

```python
#各職業的人數
prof = customers['Profession'].value_counts()
prof.plot.barh(figsize=(15,7), xlabel='Profession', title='Counting by Profession');
```

```python
#各職業的占比
prof = customers['Profession'].value_counts() / 2000
prof.plot.barh(figsize=(15,7), xlabel='Profession', title='Percents by Profession');
```
